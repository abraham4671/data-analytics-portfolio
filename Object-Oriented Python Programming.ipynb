```python
class laptop:
    def __init__(self, brand, ram):
        self.brand = brand
        self.ram = ram

    def specs(self):
        return f"{self.brand} laptop with {self.ram}gb ram"

l1 = laptop("dell", 16)
print(l1.specs())
```

```python
class bankaccount:
    def __init__(self, balance):
        self.__balance = balance  # private variable

    def deposit(self, amount):
        self.__balance += amount

    def get_balance(self):
        return self.__balance

acc = bankaccount(1000)
acc.deposit(500)
print(acc.get_balance())
```

```python
class animal:
    def speak(self):
        return "makes sound"

class cat(animal):
    def speak(self):
        return "meows"

c = cat()
print(c.speak())
```

```python
class bird:
    def sound(self):
        return "chirp"

class dog:
    def sound(self):
        return "bark"

def make_sound(creature):
    print(creature.sound())

make_sound(bird())
make_sound(dog())
```

```python
from abc import ABC, abstractmethod

class shape(ABC):
    @abstractmethod
    def area(self):
        pass

class square(shape):
    def __init__(self, side):
        self.side = side

    def area(self):
        return self.side * self.side

s = square(4)
print(s.area())
```

```python
class student:
    school = "green valley school"  # class variable

    def __init__(self, name):
        self.name = name  # instance variable

s1 = student("arun")
s2 = student("meena")

print(s1.name, s1.school)
print(s2.name, s2.school)
```

```python
class filehandler:
    def __init__(self, filename):
        self.filename = filename
        print("file opened")

    def __del__(self):
        print("file closed")

f = filehandler("data.txt")
```

```python
class vehicle:
    def start(self):
        return "vehicle starting"

class bike(vehicle):
    def start(self):
        return "bike starting with self start"

b = bike()
print(b.start())
```
