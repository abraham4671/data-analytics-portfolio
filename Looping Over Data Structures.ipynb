```python
colors = ["red", "blue", "green"]

for color in colors:
    print("i like", color)
```

```python
coordinates = (10, 20, 30)

for value in coordinates:
    print("coordinate value:", value)
```

```python
word = "code"

for letter in word:
    print(letter)
```

```python
student = {"name": "alex", "age": 21, "grade": "a"}

for key in student:
    print("key:", key)
```

```python
student = {"name": "alex", "age": 21, "grade": "a"}

for value in student.values():
    print("value:", value)
```

```python
student = {"name": "alex", "age": 21, "grade": "a"}

for key, value in student.items():
    print(key, "->", value)
```

```python
fruits = {"apple", "banana", "mango"}

for fruit in fruits:
    print("fruit:", fruit)
```

```python
for number in range(1, 6):
    print("number:", number)
```

```python
animals = ["cat", "dog", "rabbit"]

for index, animal in enumerate(animals):
    print(index, animal)
```

```python
names = ["anna", "ben", "cara"]
scores = [85, 90, 88]

for name, score in zip(names, scores):
    print(name, "scored", score)
```
