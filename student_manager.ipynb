```python
print("welcome to student manager")

student_name = input("enter student name: ")
student_age = int(input("enter age: "))

print("student:", student_name)
print("age:", student_age)
```

```python
marks = int(input("enter marks: "))

if marks >= 90:
    grade = "a"
elif marks >= 75:
    grade = "b"
elif marks >= 50:
    grade = "c"
else:
    grade = "fail"

print("grade:", grade)
```

```python
marks_list = []

for i in range(3):
    mark = int(input("enter mark: "))
    marks_list.append(mark)

print("all marks:", marks_list)
print("total:", sum(marks_list))
```

```python
student = {
    "name": student_name,
    "age": student_age,
    "marks": marks_list
}

for key, value in student.items():
    print(key, ":", value)

for mark in marks_list:
    print("mark:", mark)
```

```python
def calculate_average(marks):
    return sum(marks) / len(marks)

def check_result(avg):
    if avg >= 50:
        return "pass"
    else:
        return "fail"

average = calculate_average(marks_list)
result = check_result(average)

print("average:", average)
print("result:", result)
```

```python
class student:

    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.marks = []

    def add_mark(self, mark):
        self.marks.append(mark)

    def show_details(self):
        print("name:", self.name)
        print("age:", self.age)
        print("marks:", self.marks)
```

```python
s1 = student("rahul", 20)
s1.add_mark(80)
s1.add_mark(70)
s1.show_details()
```

```python
class college_student(student):

    def __init__(self, name, age, course):
        super().__init__(name, age)
        self.course = course

    def show_details(self):
        super().show_details()
        print("course:", self.course)
```

```python
c1 = college_student("anita", 21, "bsc computer science")
c1.add_mark(85)
c1.show_details()
```

```python
class student:

    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.marks = []

    def add_mark(self, mark):
        try:
            mark = int(mark)
            self.marks.append(mark)
        except valueerror:
            print("invalid mark")

    def calculate_average(self):
        if len(self.marks) == 0:
            return 0
        return sum(self.marks) / len(self.marks)

    def show_result(self):
        avg = self.calculate_average()
        if avg >= 50:
            print("result: pass")
        else:
            print("result: fail")

    def show_details(self):
        print("name:", self.name)
        print("age:", self.age)
        print("marks:", self.marks)
        print("average:", self.calculate_average())
        self.show_result()


name = input("enter name: ")
age = int(input("enter age: "))

s1 = student(name, age)

for i in range(3):
    mark = input("enter mark: ")
    s1.add_mark(mark)

s1.show_details()
```
