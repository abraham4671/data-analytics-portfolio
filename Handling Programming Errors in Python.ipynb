```python
try:
    number = int("hello")
except ValueError:
    print("Conversion failed! Please enter a valid number.")
```

```python
try:
    result = 10 / int("0")
except (ValueError, ZeroDivisionError) as error:
    print("An error occurred:", error)
```

```python
try:
    data = [1, 2, 3]
    print(data[5])
except IndexError:
    print("Index does not exist.")
except Exception:
    print("Something unexpected happened.")
```

```python
try:
    value = int("25")
except ValueError:
    print("Invalid input.")
else:
    print("Conversion successful! Value is:", value)
```

```python
try:
    file = open("sample.txt", "r")
    content = file.read()
except FileNotFoundError:
    print("File not found.")
finally:
    print("Execution completed.")
```

```python
age = -5

if age < 0:
    raise ValueError("Age cannot be negative.")
```

```python
class LowBalanceError(Exception):
    pass

balance = 100
withdraw = 500

if withdraw > balance:
    raise LowBalanceError("Insufficient funds!")
```
